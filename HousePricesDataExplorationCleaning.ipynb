```python
import pandas as pd

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
# Dropping sparse columns

to_drop = ['Id']
to_drop.append('Alley')
to_drop.append('MasVnrType')
to_drop.append('FireplaceQu')
to_drop.append('PoolQC')
to_drop.append('Fence')
to_drop.append('MiscFeature')

for column in to_drop:
    train.drop(column, inplace=True, axis=1)
    test.drop(column, inplace=True, axis=1)
```

```python
train.info()
```

```python
#  Label encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for column in train.columns:
    if train[column].dtype == 'O':
        train[column] = le.fit_transform(train[column])
    if column != 'SalePrice' and test[column].dtype == 'O':
        test[column] = le.fit_transform(test[column])
```

```python
for column in train.columns:
    if sum(train[column].isna()) in range(1, 300):
        train[column].fillna(train[column].mean(), inplace=True)
    if column != 'SalePrice' and sum(test[column].isna()) in range(1, 300):
        test[column].fillna(test[column].mean(), inplace=True)
```

```python
for column in train.columns:
    if train[column].dtype == 'float64':
        train[column] = train[column].astype('int64')
    if column != 'SalePrice' and test[column].dtype == 'float64':
        test[column] = test[column].astype('int64')
```

```python
train.to_csv('train_clean.csv', index=False)
test.to_csv('test_clean.csv', index=False)
```
